#### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

#### Load Dataset

```python
df = pd.read_excel(r"C:\Users\Advaith's PC\OneDrive\Desktop\DATA_ANALYSIS\spotify datasets (2).xlsx")

print(df.head(10))
```

#### Check Shape

```python
 df.shape
```

#### Dataset Information

```python
df.info()
```

#### Check Missing Values

```python
df.isnull().sum()
```

#### cleaning missing values

```python
df = df.drop(columns=['Unnamed: 13'])
```

#### checking data is cleaned or not

```python
df.isnull().sum()
```

#### remove unwanted columns

```python
df.drop(columns=['Unnamed: 14'], inplace=True)

print(df.columns)
```

#### Remove duplicates

```python
df.drop_duplicates(inplace=True)

print(df.shape)
```

#### Statistical Summary

```python
df.describe()
```

#### VISUALIZATIONS

#### Churn Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='is_churned', data=df)

plt.title("Churn Distribution")
plt.show()
```

#### Gender vs Churn

```python
plt.figure(figsize=(6,4))
sns.countplot(x='gender', hue='is_churned', data=df)

plt.title("Gender vs Churn")
plt.show()
```

#### Age Group vs Churn

```python
plt.figure(figsize=(8,5))
sns.countplot(x='age_group', hue='is_churned', data=df)

plt.title("Age Group vs Churn")
plt.xticks(rotation=45)
plt.show()
```

#### Subscription Type vs Churn

```python
plt.figure(figsize=(8,5))
sns.countplot(x='subscription_type', hue='is_churned', data=df)

plt.title("Subscription Type vs Churn")
plt.show()
```

#### Country vs Churn

```python
plt.figure(figsize=(10,5))
sns.countplot(x='country', hue='is_churned', data=df)

plt.title("Country vs Churn")
plt.xticks(rotation=45)
plt.show()
```

#### Device Type vs Churn

```python
plt.figure(figsize=(8,5))
sns.countplot(x='device_type', hue='is_churned', data=df)

plt.title("Device Type vs Churn")
plt.show()
```

#### Listening Time Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['listening_time'], bins=20, kde=True)

plt.title("Listening Time Distribution")
plt.show()
```

#### Listening Time vs Churn

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='is_churned', y='listening_time', data=df)

plt.title("Listening Time vs Churn")
plt.show()
```

#### Songs Played Per Day vs Churn

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='is_churned',
            y='songs_played_per_day',
            data=df)

plt.title("Songs Played Per Day vs Churn")
plt.show()
```

#### Skip Rate vs Churn

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='is_churned',
            y='skip_rate',
            data=df)

plt.title("Skip Rate vs Churn")
plt.show()
```

#### Ads Listened vs Churn

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='is_churned',
            y='ads_listened_per_week',
            data=df)

plt.title("Ads Listened vs Churn")
plt.show()
```

#### Offline Listening vs Churn

```python
plt.figure(figsize=(6,4))
sns.countplot(x='offline_listening',
              hue='is_churned',
              data=df)

plt.title("Offline Listening vs Churn")
plt.show()
```

#### Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

#### Top Correlations with Churn

```python
corr = numeric_df.corr()['is_churned'].sort_values(ascending=False)

print(corr)
```
